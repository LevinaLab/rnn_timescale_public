```python
import os
import glob
import pickle
import matplotlib.pyplot as plt
import numpy as np

project_root = os.path.dirname(os.path.abspath(os.path.curdir))
```

```python
results_folder = os.path.join(project_root, 'results', '7681088_2024-03-13_17-58')
```

```python
result_files = glob.glob(os.path.join(results_folder, '*'))
```

```python
result_path = result_files[0]
with open(result_path, 'rb') as f:
    result_dict = pickle.load(f)
```

```python
result_dict.keys()
```

```python
len(result_dict['trained_taus'][0])
```

```python
plt.boxplot(result_dict['trained_taus']);
plt.plot([np.average(result_dict['trained_taus'][i]) for i in range(len(result_dict['trained_taus']))], '-o')
plt.ylabel("learned tau")
plt.xlabel("module")
```
